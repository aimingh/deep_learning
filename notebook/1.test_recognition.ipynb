```python
import jetson.inference
import jetson.utils
import argparse
import matplotlib.pyplot as plt
```

```python
# load the recognition network
net = jetson.inference.imageNet("googlenet")
```

```python
samples = ["samples/black_bear.jpg", "samples/brown_bear.jpg", "samples/polar_bear.jpg"]
```

```python
for sample in samples:
    img = jetson.utils.loadImage(sample) # load an image (into shared CPU/GPU memory)
    class_idx, confidence = net.Classify(img) # classify the image
    class_desc = net.GetClassDesc(class_idx) # find the object description
    img_rander = jetson.utils.cudaToNumpy(img)
    plt.figure()
    plt.imshow(img_rander)
    plt.show()
    print("image is recognized as '{:s}' (class #{:d}) with {:f}% confidence".format(class_desc, class_idx, confidence * 100))
```
