```python
import jetson.inference
import jetson.utils
import argparse
import matplotlib.pyplot as plt
```

```python
net = jetson.inference.detectNet("ssd-mobilenet-v2")
```

```python
samples = ["samples/peds_0.jpg", "samples/peds_1.jpg", "samples/peds_2.jpg", "samples/peds_3.jpg"]
```

```python
for sample in samples:
    img = jetson.utils.loadImage(sample) # load an image (into shared CPU/GPU memory)
    detections = net.Detect(img, overlay="box,labels,conf")
    img_rander = jetson.utils.cudaToNumpy(img)
    print("detected {:d} objects in image".format(len(detections)))
    for detection in detections:
        print(detection)
    plt.figure()
    plt.imshow(img_rander)
    plt.show()
```
